```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 data = pd.read_csv('weight-height.csv')
```

```python
data.info()
data.head(2)
```

```python
X = data.Height.values   # input data
y_true = data.Weight.values # output , answer
```

```python
print(X[:5],y_true[:5])
```

```python
def line(x, w=0, b=0):   #line equation
    return x * w + b

# Cost function

def mean_square_error(y_true,y_pred):
    s = (y_true - y_pred)**2
    return np.sqrt(s.mean())
```

```python
plt.figure(figsize=(10,5))
ax1 = plt.subplot(121) # rows , columns, select column
data.plot(kind = "scatter",
       x = 'Height',
       y = 'Weight',
       title = "Height vs Weight",
       ax = ax1)

bs = [-150,-100,-50,0,50,100,150]
mses = []
for b in bs:
    y_pred = line(X, w = 2, b = b)
    mse = mean_square_error(y_true,y_pred)
    mses.append(mse)
    plt.plot(X, y_pred)
    
ax2 = plt.subplot(122)
plt.plot(bs, mses,'-d')
plt.title('Cost Function for each Line')
plt.xlabel('b')
plt.ylabel('mse')
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X.shape
```

```python
model = Sequential()
model.add(Dense(1, input_shape = (1,)))
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.8), 'mean_squared_error') # lr-->> learning rate
```

```python
model.fit(X,y_true,epochs = 45)
```

```python
y_pred_deep_learning = model.predict(X)
y_pred_deep_learning
```

```python
data.plot(kind = 'Scatter',
         x = 'Height',
         y = 'Weight',
         title = 'Height Vs Weight by Deep Learning')
plt.plot(X,y_pred_deep_learning,color = 'green', linewidth = '5')
```

```python
w, b = model.get_weights()
print(w,b)
```
