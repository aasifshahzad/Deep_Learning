```python
import pandas as pd
```

```python
melbourne_file_path = '..\data\melb_data.csv'
df = pd.read_csv(melbourne_file_path)
```

```python
df = df.dropna(axis = 0)
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Selecting Prediction Target
y = df.Price
y.head()
```

```python
# Choosing Features
df_features = ['Rooms', 'Bathroom','Landsize','Lattitude','Longtitude']
X = df[df_features]
```

```python
X.describe()
import matplotlib.pyplot as plt
df[['Rooms', 'Bathroom','Landsize','Lattitude','Longtitude','Price']].corr().plot(kind = 'bar')
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# 1>>>
df_model = DecisionTreeRegressor(random_state = 1)
# 2>>>
df_model.fit(X,y)
```

```python
# 3>>>
print('Making Decision for the following five houses:')
print(X)
print('The prediction are:')
print(df_model.predict(X))
print(y)
```

```python
print(type(df_model.predict(X)))
print(type(y))

df1 = df[['Rooms', 'Bathroom','Landsize','Lattitude','Longtitude','Price']]
df1.head()
# Comparing two numpy arrays
(y==df_model.predict(X)).all()
```

```python
# Model Validation
#error=actualâˆ’predicted
from sklearn.metrics import mean_absolute_error
y_predicted = df_model.predict(X)
mean_absolute_error = [y,y_predicted]
mean_absolute_error
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model
melbourne_model = DecisionTreeRegressor()
# Fit model
melbourne_model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = melbourne_model.predict(val_X)
mean_absolute_error =(val_y, val_predictions)
mean_absolute_error


```
